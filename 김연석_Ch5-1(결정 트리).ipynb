```python
import pandas as pd
wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine.head()
```

```python
#info()메서드 _ 데이터프레임의 각 열의 데이터 타입과 누락된 데이터가 있는지 확인하는 데 유용
```

```python
wine.info()
```

```python
#describe()메서드 _ 열에 대한 간략한 통계를 출력해 줍니다.
```

```python
wine.describe()
```

```python
data = wine[['alcohol', 'sugar', 'pH']]
target = wine['class']
```

```python
#train_test_split()은 설정값을 지정하지 않으면 25%를 테스트 세트로 지정
#여기서는 샘플 개수가 충분하므로 20% 정도만 테스트 세트로 조정
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size = 0.2, random_state = 42)
```

```python
print(train_input.shape, test_input.shape)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
print(lr.coef_, lr.intercept_)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target)
print(dt.score(train_scaled, train_target)) #훈련세트
print(dt.score(test_scaled, test_target)) #테스트세트
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=1, filled=True,
           feature_names=['alchol', 'sugar', 'pH'])
plt.show()
```

```python
#지니 불순도 = 1 - (음성 클래스 비율^2 + 양성 클래스 비율^2)
```

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(train_scaled, train_target)
print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
plt.figure(figsize=(20,15))
plot_tree(dt, filled=True, feature_names=['alchol', 'sugar', 'pH'])
plt.show()
```
