### [ 순차 데이터 ]  
: 텍스트나 시계열 데이터와 같이 순서에 의미가 있는 데이터  
: 완전 연결 신경망이나 합성곱 신경망은 이런 기억 장치 X  
: 하나의 샘플을 사용하여 정방향 계산을 수행하고 나면 그 샘플은 버려지고 다음 샘플을 처리할 때 재사용 X  


`피드포워드 신경망`: 데이터의 흠름이 앞으로만 전달되는 신경망

#### [ 순환 신경망 ]  
: 뉴런의 출력이 다시 자기 자신으로 전달(어떤 샘플을 처리할 때 바로 이전에 사용했던 데이터를 재사용)  
: 층을 `셀`이라고 부름  
: 셀의 출력을 `은닉 상태`라고 부름  
: 합성곱 신경망과 기본 구조는 같지만 층의 출력(은닉 상태)을 다음 스텝에 재사용함  
: 순환 신경망에서도 활성화 함수가 반드시 필요  
: 순환 신경망의 뉴런은 가중치가 하나 더 존재(이전 타임스텝의 은닉 상태에 곱해지는 가중치)

#### [ 셀의 가중치와 입출력 ]  
: 순환층은 일반적으로 샘플마다 `2개의 차원`을 가짐  
: 보통 하나의 샘플을 하나의 `시퀀스`라고 함  
* 시퀀스  
1. 시퀀스 안에는 여러 개의 아이템이 들어 있음  
2. 시퀀스의 길이가 바로 `타임스텝의 길이`  

: 순환층을 통과하면 1차원 배열로 변환(= 1차원 배열의 크기는 순환층의 뉴런의 개수에 의해 결정)  

* 마지막 타임 스텝의 은닉 상태에 압축하여 전달하는 것으로 볼 수 있음 ==> 그래서 순환 신경망이 정보를 기억하는 메모리를 가진다고도 표현
