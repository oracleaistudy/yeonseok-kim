#### [ 합성곱 ]  
: 입력 데이터 전체에 가중치를 적용하는 것이 아니라 일부에 가중치를 곱함  
: 연결 신경망과 달리 뉴런을 필터 혹은 커널이라고 부름  
: 여기서는 필터 = 뉴런의 개수, 커널 = 입력에 곱해지는 가중치를 의미함

특성 맵: 합성곱 계산을 통해 얻은 출력

#### [ 케라스 합성곱 층 ]

```python
import keras
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu')
```

Conv2D 클래스이 첫 번째 매개변수는 필터의 개수이다.  
kernel_size 매개변수는 필터에 사용할 커널의 크기를 지정한다.  
필터의 개수와 커널의 크기는 반드시 지정해야 하는 매개변수이다.

패딩: 입력 배열의 주의를 가상의 원소로 채우는  것  
(1) 세임 패딩: 입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 0으로 패딩하는 것  
(2) 밸리드 패딩: 패딩 없이 순수한 배열에서만 합성곱을 하여 특성 맵을 만드는 경우

```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same')
```

스트라이드: 이동의 크기  
기본으로 스트라이드는 1이다.

```python
keras.layers.Conv2D(10, kernel_size=(3,3), activation='relu', padding='same', strides=1)
```

##### [ 풀링 ]  
: 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할을 수행
: 하지만 특성 맵의 개수는 줄이지 않음  
: 차원의 개수는 유지하고 너비와 높이만 줄임  
: 가중치가 없으며 도장을 찍은 영역에서 가장 큰 값(최대 풀링)을 고르거나 평균값(평균 풀링)을 계산

```python
keras.layers.MaxPooling2D(2)
```

```python
keras.layers.MaxPooling2D(2, strides=2, padding='valid')
```
